```python
import numpy as np
import os

from tflite_model_maker.config import QuantizationConfig
from tflite_model_maker.config import ExportFormat
from tflite_model_maker import model_spec
from tflite_model_maker import object_detector

import tensorflow as tf
assert tf.__version__.startswith('2')

tf.get_logger().setLevel('ERROR')
from absl import logging
logging.set_verbosity(logging.ERROR)
```

```python
spec = model_spec.get('efficientdet_lite3')
```

```python
!unzip data.zip
```

```python
train_data = object_detector.DataLoader.from_pascal_voc('/content/train','/content/train',['Plastic','Can','Paper'])
validation_data = object_detector.DataLoader.from_pascal_voc('/content/valid','/content/valid',['Plastic','Can','Paper'])
```

```python
model = object_detector.create(train_data, model_spec=spec, epochs = 30, batch_size=8, train_whole_model=True, validation_data=validation_data)
```

```python
model.evaluate(validation_data)
```

```python
model.export(export_dir='.')
```
